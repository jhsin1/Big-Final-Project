# Safety metric

https://data.wprdc.org/dataset/police-incident-blotter/resource/1797ead8-8262-41cc-9099-cbc8a161924b

```python
import pandas as pd
import matplotlib.pyplot as plt
import array
blot = pd.read_csv("blotter.csv")
```

```python
nbh = blot['INCIDENTNEIGHBORHOOD'].unique()
print("List of neighborhoods with reported crimes: ")
print(nbh)
```

```python
hv = list()
hierdict = dict()
for x in nbh:
    if not type(x) == float:
        temp = blot['INCIDENTNEIGHBORHOOD'].str.contains(x, na=False)
        fin = blot[temp]
        temptot = fin['HIERARCHY'].sum()
        hv.append(temptot);
        hierdict.update({x : temptot})

hv.sort()
print(hv)
print(hierdict)
min = hierdict['Upper Hill']
minName = "Upper Hill"
for x in hierdict:
    if hierdict[x] < min:
        min = hierdict[x]
        minName = x
```

```python
plt.bar(hierdict.keys(), hierdict.values())
plt.show()
```

```python
N = 10
top10 = dict(sorted(hierdict.items(),key = lambda x : x[1])[:N])
print(top10)
plt.bar(top10.keys(), top10.values())
plt.xticks(rotation = 90) ## upper lawrenceville garfield - 740 - south oakland
```

```python
N = 10
top10 = dict(sorted(hierdict.items(),key = lambda x : x[1], reverse = True)[:N])
print(top10)
plt.bar(top10.keys(), top10.values())
plt.xticks(rotation = 90)
```
