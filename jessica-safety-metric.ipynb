☆ **All the imports and utility set-ups**

```python
import pandas as pd
import matplotlib.pyplot as plt
import array
blot = pd.read_csv("blotter.csv")
```

☆ **All are towards data processing, sorting, plotting, and manipulating**

```python
nbh = blot['INCIDENTNEIGHBORHOOD'].unique()
print("List of neighborhoods with reported crimes: ")
print(nbh)
```

☆ **List of neighborhoods where crimes are reported ~90**

```python
hv = list()
hierdict = dict()
for x in nbh:
    if not type(x) == float:
        temp = blot['INCIDENTNEIGHBORHOOD'].str.contains(x, na=False)
        fin = blot[temp]
        temptot = fin['HIERARCHY'].sum()
        hv.append(temptot);
        hierdict.update({x : temptot})

hv.sort()
print(hv)
```

☆ **Sorted list of the hierarchy values**

```python
print(hierdict)
```

☆ **Printed version of the dictionary with every neighborhood and their corresponding crime hierarchy value**

```python
plt.bar(hierdict.keys(), hierdict.values())
plt.show()
```

☆ **Overall view of the data**

```python
N = 20
top10 = dict(sorted(hierdict.items(),key = lambda x : x[1])[:N])
plt.bar(top10.keys(), top10.values())
plt.xticks(rotation = 90)
```

☆ **Best 20 neighborhoods - safest in terms of crime**

```python
N = 20
top10 = dict(sorted(hierdict.items(),key = lambda x : x[1], reverse = True)[:N])
plt.bar(top10.keys(), top10.values())
plt.xticks(rotation = 90)
```
