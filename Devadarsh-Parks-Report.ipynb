# Neighborhoods With the Most Amount of Parks


```python
import pandas as pd

# Reading the parks.csv file into a dataframe
park = pd.read_csv("parks.csv")

# Filtering the Dataframe to include only rows where the column 'type' only containes park
query = park['type'].str.contains('Park')
updatedParks = park[query]

# Intializing an empty parks dictionary
parksDict = {}

# Iterate over each row in the dataframe
for index, row in updatedParks.iterrows():
    neighborhoodName = row['neighborhood']
    
    if neighborhoodName in parksDict:
        parksDict[neighborhoodName] += 1 
    else:
        parksDict[neighborhoodName] = 1

# Printing the final dictionary
print(parksDict)
```

```python
# Converting to a pandas Series
parks_series = pd.Series(parksDict)

# Sorting the series in descending order
sorted_parks_series = parks_series.sort_values(ascending=False)

# Printing the sorted parks series
print(sorted_parks_series)
```

```python
import matplotlib.pyplot as plt

# Set the size of the plot for better visibility
plt.figure(figsize=(60, 15))

# Create a bar chart with keys of the dictionary on the x-axis and values on the y-axis
plt.bar(parksDict.keys(), parksDict.values())
plt.xticks(rotation=45)
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.title('Number of Parks in Each Neighborhood')
plt.grid(axis='y')

# Showing the plot
plt.show()
```

```python
# Sorting the top 20 items into a bar chart
N = 20
top20 = dict(sorted(parksDict.items(), key = lambda x : x[1], reverse = True)[:N])
print(top20)
plt.bar(top20.keys(), top20.values())
plt.xticks(rotation = 90)
```

```python
# Sorting the lowest 20 items into a bar chart
N = 20
low20 = dict(sorted(parksDict.items(), key = lambda x : x[1])[:N])
print(low20)
plt.bar(low20.keys(), low20.values())
plt.xticks(rotation = 90)
```
